```python
df[df['quantity'] > 4]
```

## Sorting Data

```python
df.sort_values('quantity',ascending= False)
```

```python
#Sorting first by quantity in ascending order then item_price in descending order
df.sort_values(['quantity','item_price'],ascending= [True, False])
```

## Grouping data

```python
#To find most ordered item in choice_description column
mst_ord=df.groupby('choice_description').sum()
mst_ord.sort_values('quantity',ascending= False).head(1)
```
